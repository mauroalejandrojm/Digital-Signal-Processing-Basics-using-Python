#### Signal Generation

```python
# Continous Signal
t = np.linspace(0, 1, 500, endpoint=False)
plt.figure()
fig, ax = plt.subplots(figsize=(16,4))
sig = np.sin(2 * np.pi * t)
plt.subplot(1, 2, 1)
plt.plot(t, sig)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('time(s)')
plt.ylabel('voltage(V)')
Fs = 5000
f = 5
sample = 1000
x = np.arange(sample)
y = np.sin(2 * np.pi * f * x / Fs)
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
# Discrete Sinusoidal Signal
t = np.linspace(0, 1, 50, endpoint=False)
plt.figure()
fig, ax = plt.subplots(figsize=(16,4))
sig = np.sin(2 * np.pi * t)
plt.subplot(1, 2, 1)
plt.stem(t, sig)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('time(s)')
plt.ylabel('voltage(V)')
Fs = 5000
f = 5
sample = 1000
x = np.arange(sample)
y = np.sin(2 * np.pi * f * x / Fs)
plt.subplot(1, 2, 2)
plt.stem(x, y)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
# Discrete Square Signal
t = np.linspace(0, 1, 50, endpoint=False)
plt.figure()
fig, ax = plt.subplots(figsize=(16,4))
sig = 2*sg.square(2*np.pi*2*t, duty=0.3)
plt.subplot(1, 2, 1)
plt.stem(t, sig)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('time(s)')
plt.ylabel('voltage(V)')
Fs = 1000
f = 5
sample = 1000
x = np.arange(sample)
y = 2*sg.square(2*np.pi*2*x/Fs, duty=0.5)
plt.subplot(1, 2, 2)
plt.stem(x, y)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
# Discrete Triangle Signal
t = np.linspace(0, 1, 50, endpoint=False)
plt.figure()
fig, ax = plt.subplots(figsize=(16,4))
sig = 2*sg.sawtooth(2*np.pi*t, width=0.2)
plt.subplot(1, 2, 1)
plt.stem(t, sig)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('time(s)')
plt.ylabel('voltage(V)')
Fs = 1000
f = 5
sample = 1000
x = np.arange(sample)
y = 2*sg.sawtooth(2*np.pi*x/Fs, width=0.5)
plt.subplot(1, 2, 2)
plt.stem(x, y)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
# Discrete Square Signal
t_ini = -1
t_fin = 1
dt = 0.001
t = np.arange(t_ini,t_fin,dt)
u0 = np.piecewise(t,t>=-0.3,[2,1])
plt.figure()
fig, ax = plt.subplots(figsize=(16,4))
sig = sg.unit_impulse(50, 'mid')
plt.subplot(1, 2, 1)
plt.plot(t, u0)
plt.xlim(-1,3)
plt.ylim(0,2.5)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('t')
plt.ylabel('escalon u(t)')
t = np.linspace(0, 1, 50, endpoint=False)
imp = 0
dt = 0.001
delta = np.piecewise(t,t>=0,[1,0])
delta_dt = np.piecewise(t,t>=imp+dt,[1,0])
unimpulso = delta - delta_dt
plt.subplot(1, 2, 2)
plt.plot(t, unimpulso)
plt.xlim(-1,1)
plt.grid(color='#000000', linestyle='-', linewidth=0.5)
plt.xlabel('time(s)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
# Shifting 

plt.figure(figsize=(16,4))
t0 = 2
t = np.arange(-1,1,0.01)
Y = np.sin(2*np.pi*t)
Y_Des = np.sin(2*np.pi*t-t0)
plt.show()
plt.figure()
plt.plot(t,Y,t,Y_Des)
plt.show()
```

#### Interactive plots in Jupyter 

```python
%matplotlib inline
from ipywidgets import interact
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

t_ini = -1
t_fin = 1
dt = 0.001
def f(t0):
    #sns.distplot(np.random.normal(1, t0, 1000))
    t = np.arange(t_ini,t_fin,dt)
    y1 = np.cos(2*np.pi*t*1/2)
    y2 = np.cos(2*np.pi*t*1/2-t0)
    plt.plot(t,y1,t,y2)
    plt.show()
interact(f, t0 = (0,10))
```
